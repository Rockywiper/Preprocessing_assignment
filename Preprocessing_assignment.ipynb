### 12.	Read the demand transaction file again and perform following operations for practice
a.	Handle the missing values in the column Percent_Quantity_With_Discount with strategy = mean


```python
from sklearn.impute import SimpleImputer
```

```python
a = SimpleImputer(missing_values = np.nan, strategy = 'mean')
```

```python
a.fit(df['Percent_Quantity_With_Discount'].values.reshape(-1,1))
df['Percent_Quantity_With_Discount'] = a.transform(df['Percent_Quantity_With_Discount'].values.reshape(-1,1))
df.head()
```
